```python
import keras
import numpy as np
from typing import List
import tensorflow as tf
import pisces.models as pm
from importlib import reload
from matplotlib import pyplot as plt
from sklearn.metrics import roc_curve, auc
from sklearn.model_selection import LeaveOneOut
from pisces.utils import avg_steps
from pisces.data_sets import DataSetObject, ModelInputSpectrogram, ModelOutputType, DataProcessor, PSGType
```

```python
data_path = "/content/drive/Shareddrives/ExternallySharedData/Pisces_fine_tuning/pre_processed_data/"
save_path = "/content/drive/Shareddrives/ExternallySharedData/Pisces_fine_tuning/fine_tuning/"
```

```python
epochs = 50
break_up_line = "=" * 20
```

# Walch

```python
dataset = "walch"
walch_preprocessed_data = np.load(f"{data_path}/{dataset}/{dataset}_preprocessed_data.npy",
                                   allow_pickle=True)
data = walch_preprocessed_data
```

```python
for train, test in LeaveOneOut().split(range(len(walch_preprocessed_data))):
    id = walch_preprocessed_data[test[0]][1]
    print(f"\n\n{break_up_line}\nTest index {test}, subject {id}")
    model = pm.MOResUNetPretrained()
    model, result = pm.run_split(
        train_indices=train,
        preprocessed_data_set=walch_preprocessed_data,
        swc=model,
        epochs=epochs,
    )
    model.tf_model.save(f'{save_path}/{dataset}/mo_fine_tuned_LOO_{id}.keras')
    np.save(f'{save_path}/{dataset}/mo_fine_tuned_LOO_{id}.npy', result)
```
